```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# reading data
data = pd.read_sas('DR1IFF_I.xpt')
```

```python
# dropping null values
data = data.dropna()
```

```python
df = pd.concat([data['DR1_020'], data['DR1ICARB']],axis=1)
```

```python
df
```

```python
plt.scatter(df['DR1_020'], df['DR1ICARB'])
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(df)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(df['DR1_020'], df['DR1ICARB'], c=kmeans.labels_, cmap='rainbow')
```
